```python
x_data=df.dropna().loc[:,['Day']].values
xx=onehot.fit_transform(x_data)
sample=len(x_data)
y_final=df.dropna().loc[:,['Total Fatalities']].values
xx=xx.reshape(sample,1)
oo=l.fit_transform(xx)
x_ones=np.ones((sample,1))
x_final=np.append(x_ones,oo,axis=1)
```

```python
new_beta=np.random.rand(8,1)
```

```python
for j in range(10):
    y_pred=np.dot(x_final,new_beta)
    derivative=-2*(np.dot(np.transpose(y_final-y_pred),x_final))/sample
    new_beta=new_beta-(0.1*derivative.T)
    cost=((y_final-y_pred)**2)/sample
```
